# What K-Nearest Neighbor?
* A supervised machine learning that solves classification problems (can be used to solve regression problems too)
* Estimating the likelihood that a data point will become a member of one group or another based on what group the data points nearest to it belong to.
* Its a lazy learning algorithm: it doesn't perform any training when you supply the training data. It does not build any model like the linear regression does
* A non-parametric method because it doesnâ€™t make any assumptions about the underlying data distribution (In linear regression, our assumption about the data is that x and y are linearly correlated). Simply put, KNN tries to determine what group a data point belongs to by looking at the data points around it. 
* It's ideal in situations where the data points are well defined or non-linear
* To measure the closeness, calculate the Euclidean distance
* KNN performs a voting mechanism
* the class with the majority vote will become the class of the data point in question.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### Generate a dataset for our next KNN

```python
from sklearn.datasets import make_blobs
```

```python
dataset = make_blobs(n_samples= 1000, n_features=2, random_state=259, centers=[(0, 0), (0, 4), (4, 0), (5, 6)])
```

* 4000 samples
* 2 features (any data point we are going to work with will be in a 2D)
* random_states: make sure we always generate the same dataset
* 4 centers means we are going to work with 4 clusters. In other words, our dataset are divided into 4 classes for the classification problem

```python
dataset
```

```python
features, target = make_blobs(n_samples= 1000, n_features=2, random_state=259, centers=[(0, 0), (0, 4), (4, 0), (5, 6)])
```

```python
features, target
```

```python
data = pd.DataFrame(features, columns=['Feature1', 'Feature2'])
```

```python
data['Target'] = target
```

```python
data.head()
```

### Plot the dataset

```python
plt.scatter(data.Feature1, data.Feature2)
```

We can make a better plot with seaborns

### Create train-test-split to avoid overfitting

Overfitting happens when our model learns the trainning dataset so well (even irrelevant noise), and turns out to work bad in the testing dataset

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[['Feature1', 'Feature2']], 
                                                    data['Target'], 
                                                    random_state=259, 
                                                    test_size=0.2, 
                                                    stratify=data['Target'])
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
y_train
```

### Create a K Nearest Neighbors model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=4, weights="distance")
classifier.fit(x_train, y_train)
```

n_neighbors: is the number of neighbors (thoses that are closest) used to determine the class/target of the unknown point
* too small n_neighbors can lead to overfitting (low bias, but high variance)
* too big n_neighbors can lead to underfitting (low variance, but high bias)
* how to pick the good k for a particular dataset will be discussed later

weights:
* uniform: all k neighbors have the same weight
* distance: a neighbor is assigned with a weight equals to the inverse of the distance => smaller distance, higher weight (intuitive, right?)

Lets predict the class of data point (2, 2)

```python
unknown = (2, 2)
```

```python
classifier.predict([(unknown)])
```

We can also get the k nearest neighbor 

```python
neighbors = classifier.kneighbors([unknown])
# the first returned array is the distance
# the second returned array is the index of that point in the dataset
neighbors
```

#### Lets visualize our unknow datapoint and its neighbors

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(x=x_train.Feature1, 
                y=x_train.Feature2, 
                hue=y_train, 
                palette=my_palette,
                markers=[',', '^', 'P', 'D'], 
                style=y_train,
                s=100)

sns.scatterplot(x=[unknown[0]], y=[unknown[0]], markers=['o'], color='red')


neighbors_x = []
neighbors_y = []

for i in neighbors[1][0]:
    print(i)
    neighbors_x.append(x_train.iloc[i, 0])
    neighbors_y.append(x_train.iloc[i, 1])

sns.scatterplot(x=neighbors_x, 
                y=neighbors_y, 
                markers=['o'], 
                fc='none',
                ec='red',
                s=200)
```

Lets go back and try with bigger K

At k=4, there is a tie, number of neighbors classified as "1" = number of neighbors classified as "0"
* How does KNN determines which class?
* Also, its apparent that the two squares are closer to the red dot.
* Since we are using "uniform", each neighbor has the same weight (not consider individual distance), it just pick either one or two classes when there is a tie (in our case the class "0"). Not super accurate, right? How about uses the "distance", means that each neighbor has a weight equals to the inverse of distance.

Change the [weight] parameter to "distance" makes the model predict the red dot belonging to class "1"

With some dataset, "uniform" is better, for some "distance" weight is better. Later we will discuss how to pick between one and another

#### Draw the decision region:
Lets explore the decision region to study how the number of k neighbors will affect the prediction performance?
(overfitting and underfitting with bad choices of k)

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plt.figure(figsize=(14, 7))
plot_decision_regions(X=x_train.to_numpy(), 
                      y=y_train.to_numpy(), 
                      clf=KNeighborsClassifier(n_neighbors=1).fit(x_train, y_train), 
                      X_highlight=np.array([unknown]), 
                      markers=",^PD",
                      colors=','.join(my_colors))
```

When K=1, decision boundary unique to each trainning set, will not perform well on the testing dataset

```python
plt.figure(figsize=(14, 7))
plot_decision_regions(X=x_train.to_numpy(), 
                      y=y_train.to_numpy(), 
                      clf=KNeighborsClassifier(n_neighbors=30).fit(x_train, y_train), 
                      X_highlight=np.array([unknown]), 
                      markers=",^PD",
                      colors=','.join(my_colors))
```

When k is large
* The decision boundary is smoothed out.
* Not easily affected by outliers
* but too high k introduces to high bias (high error)

### Optimize hyperparameters: weights and K 

Getting the error rates for each combination of weight and K

```python
from sklearn.metrics import accuracy_score
```

accuracy_score() receives y_true and y_pred and calculate the accuracy score. We can use the method to compute the error rate = 1- accuracy_score()

```python
error_rate_for_uniform_models = []
error_rate_for_distance_models = []

k_range = range(1, 30)

for k in k_range:
    clf = KNeighborsClassifier(n_neighbors=k, weights="uniform")
    clf.fit(x_train, y_train)
    y_preds = clf.predict(x_test)
    y_trues = y_test
    accuracy = accuracy_score(y_trues, y_preds)
    error_rate_for_uniform_models.append(1-accuracy)
    
    clf = KNeighborsClassifier(n_neighbors=k, weights="distance")
    clf.fit(x_train, y_train)
    y_preds = clf.predict(x_test)
    y_trues = y_test
    accuracy = accuracy_score(y_trues, y_preds)
    error_rate_for_distance_models.append(1-accuracy)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(k_range, error_rate_for_uniform_models, color="green", label="uniform", marker="o", markerfacecolor="red")
plt.plot(k_range, error_rate_for_distance_models, color="blue", label="distance", marker="o", markerfacecolor="red")
plt.legend()
plt.xlabel("K")
plt.ylabel("Error rate")
plt.show()
```

#### GridSearch to choose the best parameters for our model

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(KNeighborsClassifier(), 
                           param_grid={'n_neighbors': range(1,51), 'weights': ["uniform", "distance"]},
                          scoring="accuracy")
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
best_clf = grid_search.best_estimator_
```

#### Make the prediction on the testing set and draw confusion matrix

```python
y_test_preds = best_clf.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
sns.reset_orig() # dont want seaborn anymore
ConfusionMatrixDisplay.from_predictions(y_test, y_test_preds, cmap='magma')
```
